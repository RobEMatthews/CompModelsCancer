# Preprocess Fluorescence Spectra

## Convention

-If reshape is possible (i.e. symmetric file data, no missing files):
In the 3D matrices, each row represents spatial data, column temporal data and layered fluorescence data.

-Otherwise (missing files)*: In the 2D matrices, each row represents data in numerical order of file names (in directory) and column fluorescence data.

```python
import os
import glob

import numpy as np
import matplotlib.pyplot as plt
```

## Removal of averaged dark background from spectra:

\begin{equation}
F_{tissue}(\lambda) = F_{tissue}(\lambda) - F_{avg,bg}(\lambda)
\end{equation}

## 45_190919_QEP013392

### Background (fl_bg)

```python
files_path = r"C:\Users\Admin\Downloads\45_190919\fluo\Background"
read_files_bg = glob.glob(os.path.join(files_path,"*.txt"))

fl_bg = []
wavelen = []

for files in read_files_bg:
    with open(files) as f:
        next(f)
        next(f)
        for line in f:
            wavelen.append(float(line.split('\t')[0]))
            fl_bg.append(float(line.split('\t')[1]))

bg_len = int(len(fl_bg) / len(read_files_bg))
wavelen = wavelen[:bg_len]
fl_bg = np.array(fl_bg).reshape(-1, bg_len)
bg_mean = np.mean(fl_bg, axis = 0)  # mean along rows (axis 0)
```

### Fat (fl_fat)

```python
files_path = r"C:\Users\Admin\Downloads\45_190919\fluo\Fat"
read_files = glob.glob(os.path.join(files_path,"*.txt"))

fl_fat = []

for files in read_files:
    with open(files) as f:
        next(f)
        next(f)
        for line in f:
            fl_fat.append(float(line.split('\t')[1]))
            
fl_fat = np.array(fl_fat).reshape((len(read_files), len(wavelen)))  # convert array to matrix
fl_fat = fl_fat - bg_mean  # subtract averaged background 

fl_fat = np.reshape(fl_fat, (-1, len(read_files_bg), len(wavelen)))  # sort array spatially and temporally (if possible)

fig = plt.figure(figsize=(10,5))
plt.title("45_190919_fl_fat_QEP013392")
for i in range(len(fl_fat)):
    for j in range(len(fl_fat[0])):
        plt.plot(wavelen, fl_fat[i][j], label=("fat %d" % i))
        plt.xlabel("Wavelength $(nm)$")
        plt.ylabel("Fluorescence $(arb. units)$")
        #plt.legend()
```

```python
files_path = r"C:\Users\Admin\Downloads\50_051219\fluo\Background"
read_files_bg = glob.glob(os.path.join(files_path,"*.txt"))

fl_bg = []

for files in read_files_bg:
    with open(files) as f:
        next(f)
        next(f)
        for line in f:
            fl_bg.append(float(line.split('\t')[1]))

bg_len = int(len(fl_bg) / len(read_files_bg))
fl_bg = np.array(fl_bg).reshape(-1, bg_len)
bg_mean = np.mean(fl_bg, axis = 0)  # mean along rows (axis 0)
```

### Fat (fl_fat)*

```python
files_path = r"C:\Users\Admin\Downloads\50_051219\fluo\Fat"
read_files = glob.glob(os.path.join(files_path,"*.txt"))

fl_fat = []

for files in read_files:
    with open(files) as f:
        next(f)
        next(f)
        for line in f:
            fl_fat.append(float(line.split('\t')[1]))
            
fl_fat = np.array(fl_fat).reshape((len(read_files), len(wavelen)))
fl_fat = fl_fat - bg_mean  # subtract averaged background 

fig = plt.figure(figsize=(10,5))
plt.title("50_051219_fl_fat_QEP013392")
for i in range(len(fl_fat)):
    plt.plot(wavelen, fl_fat[i])
    plt.xlabel("Wavelength $(nm)$")
    plt.ylabel("Fluorescence $(arb. units)$")
```
